В вашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктов за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

**Описание данных**
- airports_nearest — расстояние до ближайшего аэропорта в метрах (м)
- balcony — число балконов
- ceiling_height — высота потолков (м)
- cityCenters_nearest — расстояние до центра города (м)
- days_exposition — сколько дней было размещено объявление (от публикации до снятия)
- first_day_exposition — дата публикации
- floor — этаж
- floors_total — всего этажей в доме
- is_apartment — апартаменты (булев тип)
- kitchen_area — площадь кухни в квадратных метрах (м²)
- last_price — цена на момент снятия с публикации
- living_area — жилая площадь в квадратных метрах(м²)
- locality_name — название населённого пункта
- open_plan — свободная планировка (булев тип)
- parks_around3000 — число парков в радиусе 3 км
- parks_nearest — расстояние до ближайшего парка (м)
- ponds_around3000 — число водоёмов в радиусе 3 км
- ponds_nearest — расстояние до ближайшего водоёма (м)
- rooms — число комнат
- studio — квартира-студия (булев тип)
- total_area — площадь квартиры в квадратных метрах (м²)
- total_images — число фотографий квартиры в объявлении

Пояснение: апартаменты — это нежилые помещения, не относящиеся к жилому фонду, но имеющие необходимые условия для проживания.

Для пропусков в данных о высоте потолков было принято решение заполнить данные средним значением высоты потолков для зданий схожей этажности.<br>

```python
sorted_data = df.query('ceiling_height == "NaN"')
plt.figure(figsize=(14, 5))
ax = sns.distplot(
    sorted_data['floors_count'],
    hist=True,
    bins=35,
    kde=False,
    color='red'
)

ax.set(
    xlabel='Количество этажей (шт)',
    ylabel='Количество пропущенных значений',
    title='Количество пропущенных значений в столбце "Высота потолков" при разном количестве этажей'
)
```

Заменим пропущенные значения на медианные по группе той же этажности. Для начала, посмотрим на медианы.

```python
df.groupby('floors_count')['ceiling_height'].median()
```

Как видим, нашлись ещё подозрительные значения:
- в 33х этажных домах значение медианы снова NaN.
- в 36ти этажных домах медиана высоты потолка 14.85м
- в 37ми этажных домах медиана высоты потолка 1.75м

Посмотрим на эти данные. 

```python
df.query('floors_count == 33 or floors_count == 36 or floors_count == 37')
```

Видим следующее:
- 33х этажный дом только один. Заменим пропущенное значение на медиану по всем данным
- 37ми этажный дома также только один. Заменим странное значение в 1.75м на медиану по всем данным
- нашлась запись, в которой высота потолка 27 метров. Тут явно видна ошибка с вводом или обработкой данных. Смещена запятая. Проверим всю таблицу, возможно, таких данных гораздо больше и, если это предположение подтвердится, поделим значение на 10

```python
df.query('ceiling_height > 10')
```

Обнаружилось 25 строк с ошибочно большой высотой потолков. Поделим все эти значения на 10.<br>
Снова проверим медианы высот потолков после этого. 

```python
df.loc[df['ceiling_height'] > 10, 'ceiling_height'] = df.loc[df['ceiling_height'] > 10, 'ceiling_height']/10
df.groupby('floors_count')['ceiling_height'].median()
```

Теперь заменим NaN значения 33х этажей и 37ми. Присвоим им значение медианы по всем данным, а остальные пропуски заменим при помощи функции **fillna()** на медиану по каждой группе.
Убедимся в отсутствии NaN значений столбца **ceiling_height**

```python
df.loc[df['floors_count'] == 33, 'ceiling_height'] = df['ceiling_height'].median()
df.loc[df['floors_count'] == 37, 'ceiling_height'] = df['ceiling_height'].median()
```

```python
df['ceiling_height'] = df.groupby('floors_count')['ceiling_height'].fillna(df.groupby('floors_count')['ceiling_height'].transform('median'))
df.info(memory_usage='deep')
```

```python
def draw_histogram(column, title, 
                   figsize=(14,5), color='red',
                   xlabel='', ylabel='', bins=None, xticks=0,
                   xlim=None, ylim=None):
    plt.figure(figsize=figsize)
    
    ax = sns.distplot(
        column,
        kde=False,
        hist=True,
        color=color,
        bins=bins
    )
    
    if (xlim != None):
        plt.xlim(xlim)
    if (ylim != None):
        plt.ylim(ylim)
    
    ax.set(
        xlabel=xlabel,
        ylabel=ylabel,
        title=title
    )
    
    if xticks != 0:
        ax.set(xticks=xticks)
```

```python
def draw_boxplot(column, title,
                 figsize=(15,3), color='red',
                 xlabel='', xlim=None, xticks=0):
    plt.figure(figsize=figsize)
    
    ax = sns.boxplot(
        x=column,
        color=color
    )
    
    if (xlim != None):
        plt.xlim(xlim)
    
    ax.set(
        xlabel=xlabel,
        title=title
    )
    
    if xticks != 0:
        ax.set(xticks=xticks)
```

```python
def draw_plot(data, x, y, title, function='count',
              figsize=(14,5), color='red',
              xlabel='', ylabel=''):
    
    categorised_data = data.pivot_table(index=x, values=y, aggfunc=function)
    categorised_data.columns = [y]
    categorised_data[x] = categorised_data.index
    categorised_data = categorised_data.sort_values(by='last_price')

    plt.figure(figsize=figsize)
    
    ax = sns.barplot(data=categorised_data, x=x, y=y)
    
    ax.set(
        xlabel=xlabel,
        ylabel=ylabel,
        title=title
    )
```

[Назад к оглавлению](#TOC)

Для удобства анализа зависимости стоимости квартиры от удалённости от центра, добавим в таблицу столбец, который будет содержать в себе один из типов "удалённости от центра". Посмотрев на онлайн карты, расстояние от "центра" города до его границ, определим следующие группы: 
- область: расстояние больше 20км
- окраина: расстояние от 6км до 20км
- центр города: расстояние от 2км до 6км
- самый центр города: расстояние до 2км

Напишем функцию, которая получает значение удалённости от центра и возвращает тип категории. Применим её к таблице.

```python
def get_center_distance_category(value):
    if value < 2000:
        return 'самый центр'
    elif value < 6000:
        return 'центр'
    elif value < 20000:
        return 'окраина'
    else:
        return 'область'

good_data['centers_nearest_category'] = good_data['city_centers_nearest'].apply(get_center_distance_category)
```

Напишем функцию, которая отрисовывает график по категориям. На вход передадим имя столбца, который содержит категории, по которым необходимо собрать сводную таблицу и построить график столбчатого типа.
