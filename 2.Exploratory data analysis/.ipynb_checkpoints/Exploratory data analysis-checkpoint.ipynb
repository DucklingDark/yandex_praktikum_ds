# Исследование объявлений о продаже квартир

В вашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктов за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

**Описание данных**
- airports_nearest — расстояние до ближайшего аэропорта в метрах (м)
- balcony — число балконов
- ceiling_height — высота потолков (м)
- cityCenters_nearest — расстояние до центра города (м)
- days_exposition — сколько дней было размещено объявление (от публикации до снятия)
- first_day_exposition — дата публикации
- floor — этаж
- floors_total — всего этажей в доме
- is_apartment — апартаменты (булев тип)
- kitchen_area — площадь кухни в квадратных метрах (м²)
- last_price — цена на момент снятия с публикации
- living_area — жилая площадь в квадратных метрах(м²)
- locality_name — название населённого пункта
- open_plan — свободная планировка (булев тип)
- parks_around3000 — число парков в радиусе 3 км
- parks_nearest — расстояние до ближайшего парка (м)
- ponds_around3000 — число водоёмов в радиусе 3 км
- ponds_nearest — расстояние до ближайшего водоёма (м)
- rooms — число комнат
- studio — квартира-студия (булев тип)
- total_area — площадь квартиры в квадратных метрах (м²)
- total_images — число фотографий квартиры в объявлении

Пояснение: апартаменты — это нежилые помещения, не относящиеся к жилому фонду, но имеющие необходимые условия для проживания.

#### Высота потолков

[Назад к оглавлению](#TOC)

Для пропусков в данных о высоте потолков было принято решение заполнить данные средним значением высоты потолков для зданий схожей этажности.<br>
Посмотрим при какой этажности пропущены данные. Для этого сделаем запрос к данным, в котором укажем, что хотим видеть только те данные, в которых значение высоты потолков равно **NaN**. После этого построим гистрограмму относительно столбца с количеством этажей **floors_count**.

```python
sorted_data = df.query('ceiling_height == "NaN"')
plt.figure(figsize=(14, 5))
ax = sns.distplot(
    sorted_data['floors_count'],
    hist=True,
    bins=35,
    kde=False,
    color='red'
)

ax.set(
    xlabel='Количество этажей (шт)',
    ylabel='Количество пропущенных значений',
    title='Количество пропущенных значений в столбце "Высота потолков" при разном количестве этажей'
)
```

Заменим пропущенные значения на медианные по группе той же этажности. Для начала, посмотрим на медианы.

```python
df.groupby('floors_count')['ceiling_height'].median()
```

Как видим, нашлись ещё подозрительные значения:
- в 33х этажных домах значение медианы снова NaN.
- в 36ти этажных домах медиана высоты потолка 14.85м
- в 37ми этажных домах медиана высоты потолка 1.75м

Посмотрим на эти данные. 

```python
df.query('floors_count == 33 or floors_count == 36 or floors_count == 37')
```

Видим следующее:
- 33х этажный дом только один. Заменим пропущенное значение на медиану по всем данным
- 37ми этажный дома также только один. Заменим странное значение в 1.75м на медиану по всем данным
- нашлась запись, в которой высота потолка 27 метров. Тут явно видна ошибка с вводом или обработкой данных. Смещена запятая. Проверим всю таблицу, возможно, таких данных гораздо больше и, если это предположение подтвердится, поделим значение на 10

```python
df.query('ceiling_height > 10')
```

Обнаружилось 25 строк с ошибочно большой высотой потолков. Поделим все эти значения на 10.<br>
Снова проверим медианы высот потолков после этого. 

```python
df.loc[df['ceiling_height'] > 10, 'ceiling_height'] = df.loc[df['ceiling_height'] > 10, 'ceiling_height']/10
df.groupby('floors_count')['ceiling_height'].median()
```

Теперь заменим NaN значения 33х этажей и 37ми. Присвоим им значение медианы по всем данным, а остальные пропуски заменим при помощи функции **fillna()** на медиану по каждой группе.
Убедимся в отсутствии NaN значений столбца **ceiling_height**

```python
df.loc[df['floors_count'] == 33, 'ceiling_height'] = df['ceiling_height'].median()
df.loc[df['floors_count'] == 37, 'ceiling_height'] = df['ceiling_height'].median()
```

<div class="alert alert-block alert-success">
<b>Success:</b> довольно глубокая проработка столбца, однако стоит начинать задумываться о расширяемости решений

<div class="alert alert-block alert-info">
Не совсем понял что значит "расширяемости решений".<br>
Ты про то, что таких значений может быть гораздо больше и вместо того, что бы писать руками для каждого, написать функцию для этого. Например, группируем по значениям, смотрим наличие Nan и количество таких NaN'ов, если только 1, то меняем на общую.<br>
Можешь подробнее рассказать?

<div class="alert alert-block alert-success">
Это вообще отдельная тема - промышленный подход к разработке. Она пока не очень затрагивается в курсе, но я советую тебе всегда при написании кода думать о том, насколько легко его будет применить к новому датасету, можно ли его модифицировать и не сломается ли код, когда заказчик пришлет свои данные (пусть с теми же колонками и форматами но другого содержания) за следующий год

```python
df['ceiling_height'] = df.groupby('floors_count')['ceiling_height'].fillna(df.groupby('floors_count')['ceiling_height'].transform('median'))
df.info(memory_usage='deep')
```

Добавим функции отрисовки графиков через seaborn для гистограмм и диаграмм размаха

```python
def draw_histogram(column, title, 
                   figsize=(14,5), color='red',
                   xlabel='', ylabel='', bins=None, xticks=0,
                   xlim=None, ylim=None):
    plt.figure(figsize=figsize)
    
    ax = sns.distplot(
        column,
        kde=False,
        hist=True,
        color=color,
        bins=bins
    )
    
    if (xlim != None):
        plt.xlim(xlim)
    if (ylim != None):
        plt.ylim(ylim)
    
    ax.set(
        xlabel=xlabel,
        ylabel=ylabel,
        title=title
    )
    
    if xticks != 0:
        ax.set(xticks=xticks)
```

<div class="alert alert-block alert-success">
Автоматизация это здорово!

```python
def draw_boxplot(column, title,
                 figsize=(15,3), color='red',
                 xlabel='', xlim=None, xticks=0):
    plt.figure(figsize=figsize)
    
    ax = sns.boxplot(
        x=column,
        color=color
    )
    
    if (xlim != None):
        plt.xlim(xlim)
    
    ax.set(
        xlabel=xlabel,
        title=title
    )
    
    if xticks != 0:
        ax.set(xticks=xticks)
```

Изменил изначальную функцию. Будет одна функция на все наши дальнейшие задачи<br>

```python
def draw_plot(data, x, y, title, function='count',
              figsize=(14,5), color='red',
              xlabel='', ylabel=''):
    
    categorised_data = data.pivot_table(index=x, values=y, aggfunc=function)
    categorised_data.columns = [y]
    categorised_data[x] = categorised_data.index
    categorised_data = categorised_data.sort_values(by='last_price')

    plt.figure(figsize=figsize)
    
    ax = sns.barplot(data=categorised_data, x=x, y=y)
    
    ax.set(
        xlabel=xlabel,
        ylabel=ylabel,
        title=title
    )
```

##### Удалённость от центра

[Назад к оглавлению](#TOC)

Для удобства анализа зависимости стоимости квартиры от удалённости от центра, добавим в таблицу столбец, который будет содержать в себе один из типов "удалённости от центра". Посмотрев на онлайн карты, расстояние от "центра" города до его границ, определим следующие группы: 
- область: расстояние больше 20км
- окраина: расстояние от 6км до 20км
- центр города: расстояние от 2км до 6км
- самый центр города: расстояние до 2км

Напишем функцию, которая получает значение удалённости от центра и возвращает тип категории. Применим её к таблице.

```python
def get_center_distance_category(value):
    if value < 2000:
        return 'самый центр'
    elif value < 6000:
        return 'центр'
    elif value < 20000:
        return 'окраина'
    else:
        return 'область'

good_data['centers_nearest_category'] = good_data['city_centers_nearest'].apply(get_center_distance_category)
```

У тебя есть датафрейм, и есть два слайса из него, первая половина и вторая, допустим. Ты пытаешься модифицировать один из слайсов каким-нибудь образом. Например, добавить еще одну колонку (или разделить на 1000, как в твоем примере). Но слайс может храниться в памяти не как отдельный датафрейм, а как ссылка на половину оригинального датафрейма. Тогда ты меняешь половину датафрейма, а что делать со второй половиной? Тут и возникает неопределенное поведение.
По идее, с этим можно бороться при помощи .copy. То есть заводить слайсы не как ссылки на оригинальный датафрейм, а как полноценную копию этого  датафрейма со своим выделенным местом в памяти.

Но чаще всего такие варнинги нам совершенно не страшны и мы даже рекомендуем студентам их отключать по всему проекту. Но если ты хочешь на самом деле разобраться, то вот отличная статья - https://www.machinelearningmastery.ru/understanding-settingwithcopywarning-7142952a01fa/

```python
spb_aparts = good_data.query('locality_name == "Санкт-Петербург"')
spb_aparts.loc[:, 'kms_to_center'] = (spb_aparts.loc[:, 'city_centers_nearest']/1000).round()
spb_aparts_pivot = spb_aparts.pivot_table(index='kms_to_center', values='last_price', aggfunc='median')
spb_aparts_pivot['kms_to_center'] = spb_aparts_pivot.index

plt.figure(figsize=(14,4))
ax = sns.barplot(data=spb_aparts, x='kms_to_center', y='last_price')

ax.set(
    xlabel='Удалённость от центра (километры)',
    ylabel='Цена (тысячи)',
    title='Цены на квартиры в зависимости от удалённости от центра Санкт-Петербурга'
)
```
