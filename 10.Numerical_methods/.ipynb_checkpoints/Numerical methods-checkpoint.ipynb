## Подготовка данных

Добавим класс `Helper`, который будет содержать в себе вспомогательные функции

```python
class Helper:    
    def read_and_print_info(self, path):
        df = pd.read_csv(path)
        self.print_random_rows(df, 25)
        self.print_describe(df)
        self.print_info(df)
        self.print_nans(df)
        self.print_duplicated(df)
        return df
    
    def print_random_rows(self, df, n):
        print('----- Данные -----')
        display(df.sample(n=n, random_state=SEED))
    
    def print_describe(self, df):
        print('----- Статистика о данных -----')
        display(df.describe().T)
        
    def print_info(self, df):
        print('----- Типы данных -----')
        df.info(memory_usage='deep')
        
    def print_nans(self, df):
        print('----- NaN значения -----')
        nans_info = pd.DataFrame()
        nans_info['Количество'] = df.isnull().sum()
        nans_info['Процент от общего количества'] = (nans_info['Количество'] / df.shape[0] * 100).round(2)
        nans_info = nans_info.sort_values(by='Количество', ascending=False)
        display(nans_info)
        
    def print_duplicated(self, df):
        print('----- Дубликаты -----')
        print('Дубликатов:', df.duplicated().sum())
        
    def print_value_counts(self, df, column):
        print('----- Уникальные значения -----')
        display(df[column].value_counts())
        
    def draw_boxplot(self, df, column, title, 
                     figsize=(15,3), color='red', 
                     xlabel='', xlim=None):
        plt.figure(figsize=figsize)

        ax = sns.boxplot(x=df[column], color=color)

        if (xlim != None):
            plt.xlim(xlim)

        ax.set(xlabel=xlabel, title=title)
        
    def draw_histogram(self, df, column, title, 
                       figsize=(14,5), color='red',
                       xlabel='', ylabel='', kde=False,
                       bins=None, xticks=0, xlim=None, ylim=None):
        plt.figure(figsize=figsize)

        ax = sns.distplot(df[column], color=color, kde=kde, bins=bins)

        if (xlim != None):
            plt.xlim(xlim)
        if (ylim != None):
            plt.ylim(ylim)

        ax.set(xlabel=xlabel, ylabel=ylabel, title=title)

        if xticks != 0:
            ax.set(xticks=xticks)
            
    def train_valid_test_split(self, X, y):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, 
                                                            random_state=SEED)
        X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.25, 
                                                              random_state=SEED)
        return X_train, X_valid, X_test, y_train, y_valid, y_test
    
    def elements_percent(self, title, first, second):
        print(title, '{:.0%}'.format(first.shape[0]/second.shape[0]))
        
    def all_samples_percent(self, frame, train, valid, test):
        self.elements_percent('Количество элементов обучающей выборки:', train, frame)
        self.elements_percent('Количество элементов валидационной выборки:', valid, frame)
        self.elements_percent('Количество элементов тестовой выборки:', test, frame)
        
    def features_target_split(self, df, target_column):
        X = df.drop(target_column, axis=1)
        y = df[target_column]
        return X, y
    
    def rmse(self, y, predictions):
        return sqrt(mean_squared_error(y, predictions))
```

Добавим объект созданного класса

```python
helper = Helper()
```

- столбец NumberOfPictures весь нулевой, возможно его стоит удалить
- некоторые столбцы не играют сильной роли при формировании стоимости на машину. Можно удалить
- заменим типы данных на более подходящие
- в данных есть пропуски, посмотрим на них подробнее
- присутствуют явные аномалии в данных (например, максимальное значение Power равное 2000, явный выброс)

Удалим дубликаты данных

### Удаление данных

В таблице есть данные, которые мы удалим:
- дата скачивания анкеты из базы - бесполезная ифнормация при расчёте стоимости машины
- месяц регистрации авто обычно не особо влияет на стоимость машины, в отличие от года регистрации 
- дата создания анкеты - точно также, как и с датой скачивания анкеты, бесполезные для расчёта стоиости машины данные
- количество фотографий - во всех записях количество фотографий равно 0, кроме того, количество фотографий на цену автомобиля не повлияет (на продаваемость, может быть, но не на цену)
- почтовый индекс владельца - тоже информация, которая не влияет на цену
- дата последней активности пользователя - аналогично

```python
df.drop(['date_crawled', 'registration_month', 'date_created', 'number_of_pictures', 
         'postal_code', 'last_seen'], axis=1, inplace=True)
df.columns
```

### Смена типов данных

Посмотрим ещё раз на основную статистику по данным и на типы данных

```python
helper.print_describe(df)
helper.print_info(df)
```

Заменим:
- `price` на uint16 - максимальное значение поля 20000 (и похоже, что оно аномальное), 16ти разраядного целого числа хватит
- `registrasion_year` - также, на uint16 - и тут тоже есть выброс (9999 год), отфильтруем дальше
- `power` - посмотрим на все данные, очень странно выглядит максимальное значение в 20000 - скорее всего это аномалия, но поменяем тип данных на uint16 (uint8 нам не хватит, так как значения даже в 1000 вполне реальны и не будут аномалией)
- `kilometer` - тут заменим на uint32

```python
df['price'] = df['price'].astype(np.uint16)
df['registration_year'] = df['registration_year'].astype(np.uint16)
df['power'] = df['power'].astype(np.uint16)
df['kilometer'] = df['kilometer'].astype(np.uint32)
helper.print_describe(df)
helper.print_info(df)
```

После удаления ненужных данных и смены типов данных на более подходящие, размер памяти уменьшился с 217 Мб до 127.1 Мб

```python
def add_table_row(title='', parameters='', 
                  fit_time='', pred_time='', 
                  rmse=''):
    top.append({'Модель': title, 'Гиперпараметры': parameters, 
                'Время обучения': fit_time, 'Время предсказаний': pred_time, 
                'RMSE': rmse})
```

Теперь проверим разные классификаторы на разных выборках:
- дамми классификатор - используем его для проверки моделей на адекватность, скормим ему данные без модели машины, после кодирования признаков
- линейная регрессия - проверим на данных после ordinal кодирования и без модели машины, после кодирования признаков
- LightGBM и CatBoost - им мы скормим полные данные, данные после ordinal кодирования (так сравним как модели работают при изначально категориальных данных и уже переведённых в числа данных) и данные без модели машины до кодирования данных, то есть с категориальными данными (чтобы сравнить время и качество работы моделей без модели машины и с ней)

#### LightGBM

Перейдём к градиентному бустингу, начнём с LightGBM. Использовать будем с разным значением максимальной глубины `max_depth` и количеством оценщиков `n_estimators`<br>
Проверять будет на полных данных, полных данных после `Ordinal` кодирования, на данных без модели машины. Выделим для этого категориальные признаки

```python
frames = [X_full_train, X_full_valid, X_full_test]
columns = ['vehicle_type', 'gearbox', 'model', 'fuel_type', 'brand', 'not_repaired']

for frame in frames:
    for column in columns:
        frame[column] = frame[column].astype('category')
```

```python
def LGBM_fit_predict(title, X_train, y_train, X_valid, y_valid):
    for depth in range(9, 13, 3):
        top_rmse = 10000
        top_fit_time = 0
        top_prediction_time = 0
        top_est = 0
        top_parameters = ''

        for est in range(50, 111, 60):
            lgbm = LGBMRegressor(max_depth=depth, n_estimators=est, random_state=SEED)
            
            start = time.time()
            lgbm.fit(X_train, y_train)
            fit_time = time.time() - start
            
            start = time.time()
            predictions = lgbm.predict(X_valid)
            prediction_time = time.time() - start
            
            rmse_value = helper.rmse(y_valid, predictions)
            print('Глубина', depth, '- оценщиков' , est, '- RMSE =', rmse_value, 
                  '- обучение', fit_time, '- предсказание', prediction_time)

            if (top_rmse > rmse_value):
                top_rmse = rmse_value
                top_fit_time = fit_time
                top_prediction_time = prediction_time
                top_est = est
                top_parameters = 'Глубина', depth, 'Оценщиков', est

        print('Наилучший результат при', depth, 'глубине, при', est, 'оценщиках:', top_rmse, '\n')
        add_table_row(title, top_parameters, top_fit_time, top_prediction_time, top_rmse)
```

```python
add_table_row()
LGBM_fit_predict('LGBM (Full)', 
                 X_full_train, y_full_train,
                 X_full_valid, y_full_valid)
display(pd.DataFrame(top))
```

LGBM на данный момент показал лучшие результаты по метрике RMSE. Рассмотрим результаты на каждой из выборок:

- LGBM на полных данных после Ordinal кодирования лучшее время обучения и предсказания, точность предсказний на втором месте
- LGBM на данных без модели машины показал худшую метрику RMSE

#### CatBoost

Перейдём к модели CatBoost: сделаем для неё те же действия, что и для LGBM

```python
def CatBoost_fit_predict(title, X_train, y_train, X_valid, y_valid, categories=None):
    for depth in range(9, 13, 3):
        top_rmse = 10000
        top_fit_time = 0
        top_prediction_time = 0
        top_est = 0
        top_parameters = ''

        for iteration in range(50, 111, 60):
            catboost = CatBoostRegressor(depth=depth, iterations=iteration, 
                                         cat_features=categories, random_state=SEED)
            
            start = time.time()
            catboost.fit(X_train, y_train)
            fit_time = time.time() - start
            
            start = time.time()
            predictions = catboost.predict(X_valid)
            prediction_time = time.time() - start
            
            rmse_value = helper.rmse(y_valid, predictions)
            print('Глубина', depth, '- итераций' , iteration, '- RMSE =', rmse_value, 
                  '- обучение', fit_time, '- предсказание', prediction_time)

            if (top_rmse > rmse_value):
                top_rmse = rmse_value
                top_fit_time = fit_time
                top_prediction_time = prediction_time
                top_iterations = iteration
                top_parameters = 'Глубина', depth, 'Итераций', iteration

        print('Наилучший результат при', depth, 'глубине, при', top_iterations, 'итерациях:', top_rmse, '\n')
        add_table_row(title, top_parameters, top_fit_time, top_prediction_time, top_rmse)
```

```python
add_table_row()
categories = ['vehicle_type', 'gearbox', 'model', 'fuel_type', 'brand', 'not_repaired']
CatBoost_fit_predict('CatBoost (Full)', 
                     X_full_train, y_full_train, 
                     X_full_valid, y_full_valid, 
                     categories=categories)
display(pd.DataFrame(top))
```

```python
add_table_row()
CatBoost_fit_predict('CatBoost (Ordinal)', 
                     X_full_ord_train, y_full_ord_train, 
                     X_full_ord_valid, y_full_ord_valid)
display(pd.DataFrame(top))
```

```python
add_table_row()
categories = ['vehicle_type', 'gearbox', 'fuel_type', 'brand', 'not_repaired']
CatBoost_fit_predict('CatBoost (w/o model)', 
                     X_wo_model_train, y_wo_model_train, 
                     X_wo_model_valid, y_wo_model_valid, 
                     categories=categories)
display(pd.DataFrame(top))
```

CatBoost, в сравнении с LGBM показал худшую скорость обучения, но лучшую скорость предсказаний. Рассмотрим результаты по всем выборкам:
- на полной выборке - лучшая метрика RMSE
- на выборке после Ordinal кодирования самое быстрое время обучения и предсказаний. Точность предсказаний на втором месте
- на выборке без модели машины CatBoost на втором месте по времени работы, но худшая точность предсказаний

```python
def feature_importances(model, X, title):
    feature_imp = pd.DataFrame(sorted(zip(model.feature_importances_, 
                                          X.columns)), 
                               columns=['Value','Feature'])
    plt.figure(figsize=(20, 10))
    sns.barplot(x="Value", y="Feature", data=feature_imp.sort_values(by="Value", ascending=False))
    plt.title(title)
    plt.tight_layout()
    plt.show()
```

## Вывод

Подводя итог, можно сказать следующее: 
- наилучшие модели LGBM и CatBoost, они показали наименьшиую RMSE
- время обучения у LGBM меньше
- время предсказаний у CatBoost меньше
- точность предсказаний у LGBM лучше

Порекомендовать конкретную модель мы не можем. Следует уточнить у заказчика, какой именно параметр ему важнее:
- если важнее скорость обучения и точность, то стоит использовать LGBM
- если важнее скорость предсказаний, а скорость обучения не так важна, то можно использовать CatBoost, подобрав гиперпараметры под нужную точность
- если важнее скорость предсказаний и скорость обучения, то лучше выбрать LGBM

Стоит отметить, что в определённый момент, смена гиперпараметров не влечёт за собой сильного изменения точности предсказаний, возможно стоит остановиться на достаточной точности предсказаний, но сократить время на обучение для CatBoost<br>
Также необходимо учесть, что предсказания точнее, если обучать модели на данных, которые содержат модель машины
