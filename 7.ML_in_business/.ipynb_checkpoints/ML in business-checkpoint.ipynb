# Входные данные
Информация о трёх регионах, которая содержит в себе информацию об идентификаторе места, некоторые признаки и объм запаса сырья

# Задача
Необходимо построить модель машинного обучения, которая поможет определить регион, где добыча принесёт лучшую прибыль, а также проанализировать возможные прибыль и риски

Подключим необходимые библиотеки

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from numpy.random import RandomState

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
sns.set(style="darkgrid")
```

Создадим переменную SEED для использования в качестве random_state параметра

```python
SEED = 69
```

## Загрузка и подготовка данных

Напишем функцию, которая загружает файл, выводит 20 случайных строк, основую информацию о данных и возращает датафрейм

```python
def read_and_print_info(path):
    df = pd.read_csv(path)
    display(df.sample(n=20, random_state=SEED))
    display(df.info(memory_usage='deep'))
    display(df.describe().T)
    print('Дубликатов:', df.duplicated().sum())
    return df
```

### Загрузка и первичный осмотр данных

Загрузим данные и посмотрим 20 случайных строк файлов

```python
data_zero = read_and_print_info('/datasets/geo_data_0.csv')
```

```python
data_one = read_and_print_info('/datasets/geo_data_1.csv')
```

```python
data_two = read_and_print_info('/datasets/geo_data_2.csv')
```

### Удаления данных

В данных удалим столбец `id`, так как эти данные не нужны модели машинного обучения. Нас интересуют регионы, а не конкретные скважины

```python
data_zero.drop(['id'], axis=1, inplace=True)
data_one.drop(['id'], axis=1, inplace=True)
data_two.drop(['id'], axis=1, inplace=True)

data_zero = data_zero.reset_index(drop=True)
data_one = data_one.reset_index(drop=True)
data_two = data_two.reset_index(drop=True)
```

### Распределения данных

Посмотрим на распределения данных. Добавим функции:
- две функции строят гистограмму и диаграмму размаха по полученному датафрейму и имени столбца
- ещё одна функция принимает на вход датафрейм и передаёт его первым двум функции вместе с именами столбцов

```python
def draw_hist(column, subtitle, ax):
    sns.distplot(column, bins=20, color='red', ax=ax)
    ax.set(title=subtitle)
```

```python
def draw_box(column, subtitle, ax):
    sns.boxplot(y=column, color='red', ax=ax)
    ax.set(title=subtitle)
```

```python
def draw_distribution(df, title):
    for column in df.columns:
        fig, axes = plt.subplots(1, 2, figsize=(16, 6))
        fig.suptitle(title)
        draw_hist(df[column], ('Распределение ' + column), axes[0])
        draw_box(df[column], ('Диаграмма размаха ' + column), axes[1])
```

Посмотрим на регион `Zero`

Теперь регион `One`

```python
draw_distribution(data_one, title='Регион Zero')
```

И регион `Two`

```python
draw_distribution(data_two, title='Регион Zero')
```

Так как мы не знаем, что значат признаки и их значения, нам сложно оценить адекватность распределений и этих выбросов. **Оставим данные в исходном виде**

## Обучение и проверка модели

### Выделение целевого признака

Выделим целевой признак

```python
features = ['f0', 'f1', 'f2']
target = ['product']

X_0 = data_zero[features]
y_0 = data_zero[target]
X_1 = data_one[features]
y_1 = data_one[target]
X_2 = data_two[features]
y_2 = data_two[target]
```

### Разбиение данных, обучение и проверка модели

Добавим функцию, которая выводит процент количества элементов одного датафрейма от другого

```python
def elements_percent(title, first, second):
    print(title, '{:.2%}'.format(first.shape[0]/second.shape[0]))
```

Напишем функцию, которая будет делать следующее:
- на вход подадим признаки и целевой признак (X и y)
- разобъём все данные на обучающую и валидационную выборки в соотношении 3:1 или 75%:25% (напечатаем на экране соотношение получившихся выборок)
- обучим модель на обучающей выборке
- сделаем предсказания на валидационной выборке
- напечатаем средний предсказанный запас сырья и RMSE
- вернём предсказания модели и правильные ответы валидационной выборки, чтотбы сохранить их в отдельной переменной

```python
def fit_predcit(X, y):
    X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.25, random_state=SEED)
    elements_percent('Размер обучающей выборки:', X_train, X)
    elements_percent('Размер валидационной выборки:', X_valid, X)
    model = LinearRegression()
    model.fit(X_train, y_train)
    predicts = model.predict(X_valid)
    predicts_mean = predicts.mean()
    rmse = mean_squared_error(y_valid, predicts)**.5
    print('----------')
    print('Cредний предсказанный запас сырья: {:.4f}'.format(predicts_mean))
    print('RMSE модели: {:.2f}'.format(rmse))
    return y_valid, predicts
```

Проверим регион `Zero`

```python
y_valid_0, predicts_0 = fit_predcit(X_0, y_0)
```

Регион `One`

```python
y_valid_1, predicts_1 = fit_predcit(X_1, y_1)
```

Регион `Two`

```python
y_valid_2, predicts_2 = fit_predcit(X_2, y_2)
```

### Вывод

Можем сделать следующие выводы:
- в регионах `Zero` и `Two` достаточно большой показатель среднего предсказанного запаса сырья (**92.5503 и 94.9081**), но у этих регионов достаточно высокая RMSE (**37.71 и 40.21**), что говорит о том, что модель неоднозначна и не отличается высокой точностью
- в регионе `One` средний предсказанный запас сырья меньше - **69.0004**, но RMSE равен **0.89**. Можем сделать вывод, что эта модель точнее

## Подготовка к расчёту прибыли

Рассчитаем объём сырья, который необходим для безубыточной разработки новой скважины

```python
BUDGET = 10e9
BEST_WELLS_COUNT = 200
WELLS_COUNT = 500
BARELL_1K_INCOME = 4.5*10e4
```

```python
minimum_prod_value = BUDGET / (BEST_WELLS_COUNT * BARELL_1K_INCOME)
print('Объём продукта с одной скважины, необходимый для безубыточной разработки: {:.2f}'.format(minimum_prod_value))
```

### Вывод

Минимальное значение продукта с одной скважины, необходимое для безубыточной разработки - **111.11**. Такой показатель расчитан при условии, что учитываются 200 лучших скважин из выборки в 500 скважин<br>
Такое значение выше среднего значения выборок в каждом из регионов, тем не менее, в регионах есть и значения выше

## Расчёт прибыли и рисков 

Переведём данные в формат pd.Series

```python
valid_y_0 = pd.Series(y_valid_0['product']).reset_index(drop=True)
valid_y_1 = pd.Series(y_valid_1['product']).reset_index(drop=True)
valid_y_2 = pd.Series(y_valid_2['product']).reset_index(drop=True)

predicts_y_0 = pd.Series(predicts_0[:, 0])
predicts_y_1 = pd.Series(predicts_1[:, 0])
predicts_y_2 = pd.Series(predicts_2[:, 0])
```

### Расчёт прибыли

Напишем функцию для расчёта прибыли:
- выберем случайно 500 точек и отберём 200 с максимальным показателем объёма сырья
- просуммируем целевое значение объёма сырья
- рассчитаем прибыль 

```python
def income_predict(y_target, y_predict, state, title='', need_to_print=False):
    sample = y_predict.sample(n=WELLS_COUNT, replace=True, random_state=state)
    top_predictions = sample.sort_values(ascending=False).head(BEST_WELLS_COUNT)
    top_targets = y_target[top_predictions.index]
    top_mean = top_targets.mean()
    top_sum = top_targets.sum()
    top_income = top_sum * BARELL_1K_INCOME - BUDGET
    if need_to_print:
        print('-----', title, '-----')
        print('Средний запас сырья: {:.2f}'.format(top_mean))
        print('Суммарный объём сырья: {:.1f}'.format(top_sum))
        print('Прибыль для полученного суммарного объёма сырья: {:.0f}'.format(top_income))
    return top_income
```

```python
income_0 = income_predict(valid_y_0, predicts_y_0, SEED, title='Регион "Zero"', need_to_print=True)
```

```python
income_1 = income_predict(valid_y_1, predicts_y_1, SEED, title='Регион "One"', need_to_print=True)
```

```python
income_2 = income_predict(valid_y_2, predicts_y_2, SEED, title='Регион "Two"', need_to_print=True)
```

### Расчёт рисков

Добавим некоторые переменные для техники `Bootstrap`

```python
samples_count = 1000
alpha = .05
```

Теперь добавим функцию, в которой реализуем технику `Bootstrap`

```python
def risk_predict(title, y_target, y_predict):
    incomes = []
    loss_count = 0
    state = np.random.RandomState(SEED)
    for i in range(samples_count):
        income = income_predict(y_target, y_predict, state)
        incomes.append(income)
        if income < 0:
            loss_count += 1
    incomes = pd.Series(incomes)
    incomes_mean = incomes.mean()
    lower = incomes.quantile(alpha / 2)
    upper = incomes.quantile(1 - alpha / 2)
    loss_rate = loss_count / samples_count
    print('-----', title, '-----')
    print('Количество повторений Bootstrap техники:', samples_count)
    print('Средняя прибыль: {:.2f}'.format(incomes_mean))
    print('Нижняя граница доверительного интервала: {:.2f}'.format(lower))
    print('Верхняя граница доверительного интервала: {:.2f}'.format(upper))
    print('Риск убытков: {:.2%}'.format(loss_rate))
```

```python
risk_predict('Регион "Zero"', valid_y_0, predicts_y_0)
```

```python
risk_predict('Регион "One"', valid_y_1, predicts_y_1)
```

```python
risk_predict('Регион "Two"', valid_y_2, predicts_y_2)
```

### Вывод

При расчёте прибыли по 200 лучшим скважинам из 500 случайных мы получили следующее:
- во всех регионах средний запас сырья больше необходимого минимума (111.11), а именно 116.11 (`'Zero'`), 116.20 (`'One'`) и 115.0 (`'Two'`)

При применении техники `Bootstrap` результаты получены следующие:
- у региона `One` самая высокая средняя прибыль: 460 684 591.34
- только у региона `One` в 95%-й доверительный интервал не попадают убытки, а только прибыль
- только регион `One` обеспечивает риск убытков ниже **2.5%**, а именно **1.20%**
- можно рекомендовать к разработке регион `One`
