```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import re

from sklearn.metrics import roc_auc_score, accuracy_score, roc_curve
from sklearn.model_selection import train_test_split, cross_val_score, cross_validate
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from lightgbm import LGBMClassifier

from datetime import timedelta

sns.set(style="darkgrid")
warnings.filterwarnings('ignore')
SEED = 66
```

Добавим класс - помощник, который будет содержать вспомогательные функции для дальнейшей работы

```python
class Helper:    
    def read_and_print_info(self, path):
        df = pd.read_csv(path)
        self.print_random_rows(df, 25)
        self.print_describe(df)
        self.print_info(df)
        self.print_nans(df)
        self.print_duplicated(df)
        return df
    
    def print_random_rows(self, df, n):
        print('----- Данные -----')
        display(df.sample(n=n, random_state=SEED))
    
    def print_describe(self, df):
        print('----- Статистика о данных -----')
        display(df.describe().T)
        
    def print_info(self, df):
        print('----- Типы данных -----')
        df.info(memory_usage='deep')
        
    def print_nans(self, df):
        print('----- NaN значения -----')
        nans_info = pd.DataFrame()
        nans_info['Количество'] = df.isnull().sum()
        nans_info['Процент от общего количества'] = (nans_info['Количество'] / df.shape[0] * 100).round(2)
        nans_info = nans_info.sort_values(by='Количество', ascending=False)
        display(nans_info)
        
    def print_duplicated(self, df):
        print('----- Дубликаты -----')
        print('Дубликатов:', df.duplicated().sum())
        
    def print_value_counts(self, df, column):
        print('----- Уникальные значения -----')
        print('Уникальных значений:', df[column].value_counts().sum())
        print('Записей:', len(df[column]))
        
    def print_unique_values(self, df, column):
        print('----- Уникальные значения столбца', column, '-----')
        print(df[column].unique())
    
    def draw_countplot(self, df, x, title,
                       hue='exited', figsize=(14,5),
                       xlabel='', ylabel='', xticks=None):
        plt.figure(figsize=(14, 6))
        ax = sns.countplot(data=df_total, x=x, hue=hue)
        ax.set(xlabel=xlabel, ylabel=ylabel, title=title)
        
        if (xticks!=None):
            plt.xticks(range(len(xticks)), xticks)
    
    def elements_percent(self, title, first, second):
        print(title, '{:.2%}'.format(first.shape[0]/second.shape[0]))
    
    def train_test_split(self, df, target, test_size, output=True):
        X = df.drop(target, axis=1)
        y = df[target]
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=SEED)
        if (output):
            self.elements_percent('Количество элементов обучающей выборки:', 
                                  X_train, df)
            self.elements_percent('Количество элементов тестовой выборки:', 
                                  X_test, df)
            self.elements_percent('Количество элементов одного класса обучающей выборки:', 
                             y_train[y_train == 1], y_train)
            self.elements_percent('Количество элементов одного класса тестовой выборки:', 
                             y_test[y_test == 1], y_test)
        
        return X_train, X_test, y_train, y_test
    
    def feature_importances(self, model, X, title):
        feature_imp = pd.DataFrame(sorted(zip(model.feature_importances_, 
                                              X.columns)), 
                                   columns=['Value','Feature'])
        plt.figure(figsize=(20, 10))
        sns.barplot(x="Value", y="Feature", data=feature_imp.sort_values(by="Value", ascending=False))
        plt.title(title)
        plt.tight_layout()
        plt.show()
        
    def cross_val(self, n_model, n_X_train, n_y_train, cv=5):
        roc_auc= cross_val_score(n_model, n_X_train, n_y_train, cv=cv, scoring='roc_auc').mean()
        print(f"roc_auc: {roc_auc:.4f}")
        accuracy = cross_val_score(n_model, n_X_train, n_y_train, cv=cv, scoring='accuracy').mean()
        print(f"accuracy: {accuracy:.4f}")
```

Начнём с пола клиента

```python
def remove_and_validate(df, columns):
    new_df = df.drop(columns, axis=1)
    X_train, X_test, y_train, y_test = helper.train_test_split(new_df, 'exited', test_size=0.25, output=False)
    print('--- LGBM ---')
    model = LGBMClassifier(random_state=SEED)
    helper.cross_val(model, X_train, y_train)
    print('--- CatBoost ---')
    model = CatBoostClassifier(verbose=0, random_state=SEED)
    helper.cross_val(model, X_train, y_train)
```

```python
remove_and_validate(df_one_hot, ['senior_citizen'])
```

Метрики улучшились для `CatBoost`, а для `LGBM` уменьшились<br>
Попробуем теперь уберём данные о том, пользуется ли клиент подключением по оптоволокну

Удаление данных о типе подключения снизили точность<br>
Попробуем объединить удаление данных о поле клиента и о том, пенсионер ли он
