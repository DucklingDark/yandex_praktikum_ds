# Задача
Необходимо построить модель, которая будет предсказывать спрос на следующий час по историческим данным

Подключим необходимые библиотеки

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

from lightgbm import LGBMRegressor

import time
from math import sqrt

import warnings
warnings.filterwarnings("ignore")
```

```python
sns.set(style="darkgrid")
```

```python
SEED = 66
```

Добавим класс `ML_data` (в нём будем хранить фреймы с признаками и целевым признаком и некоторые функции) и класс `Helper` (в нём будут необходимые работы функции)

```python
class ML_data:
    def __init__(self, data: pd.DataFrame):
        self.df = data
        self.X = pd.DataFrame()
        self.y = pd.DataFrame()
        
    def x_y_split(self, column):
        self.X = self.df.drop(column, axis=1)
        self.y = self.df[column]
```

```python
class Helper:
    def __init__(self):
        pass
    
    def draw_lineplot(self, x, y, title, 
                      figsize=(14,5), color='red',
                      xlabel='', ylabel=''):
        plt.figure(figsize=figsize)
        ax = sns.lineplot(x=x, y=y, color=color)
        ax.set(xlabel=xlabel, 
               ylabel=ylabel, 
               title=title)
        
    def draw_seasonal(self, df, color='red', figsize=(14,5), 
                      need_trend=True, need_seasonal=True, 
                      need_resid=True):
        decomposed = seasonal_decompose(df)
        if need_trend:
            self.draw_lineplot(decomposed.trend.index, 
                               decomposed.trend, 'Тренд',
                               xlabel='Дата', ylabel='Количество заказов',
                               figsize=figsize)
        if need_seasonal:
            self.draw_lineplot(decomposed.seasonal.index, 
                               decomposed.seasonal, 'Сезонность', 
                               xlabel='Дата', ylabel='Количество заказов',
                               figsize=figsize)
        if need_resid:
            self.draw_lineplot(decomposed.resid.index, 
                               decomposed.resid, 'Остаток', 
                               xlabel='Дата', ylabel='Количество заказов',
                               figsize=figsize)
            
    def elements_percent(self, title, first, second):
        print(title, '{:.0%}'.format(first.shape[0]/second.shape[0]))
        
    def rmse(self, y, predictions):
        return sqrt(mean_squared_error(y, predictions))
    
    def fit_and_predict(self, model, train: ML_data, test: ML_data):
        start = time.time()
        model.fit(train.X, train.y)
        fit_time = time.time() - start
        start = time.time()
        predictions = model.predict(test.X)
        prediction_time = time.time() - start
        rmse_value = self.rmse(test.y, predictions)
        predicted = pd.Series(data=predictions, index=test.df.index)
        return fit_time, prediction_time, rmse_value, predicted
        
helper = Helper()
```

## Обучение и тестирование моделей

### Добавление признаков

Добавим функцию для создания признаков<br>
В качестве календарных признаков возьмём следующие:
- месяц и день (может помочь предсказать спрос в праздники)
- день недели (поможет с рабочими и выходными днями)
- час

```python
def make_features(df, column, max_lag, rolling_mean_size):
    data = df.copy()
    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data[column].shift(lag)

    data['rolling_mean'] = data[column].shift().rolling(rolling_mean_size).mean()
    
    return data
```

Теперь запишем данные с новыми признаками в новую переменную

```python
df_new_features = make_features(df, column='num_orders', 
                                max_lag=10, 
                                rolling_mean_size=10)
df_new_features.head(15)
```

### Разделение данных

Напишем функцию, которая будет разделять датафрейм на обучающую и тестовую выборки, а возвращать объекты класса `ML_data`. Вместе с тем, выведем на экран процент выборок от общей выборки, удалим пропуски и разделим на признаки и целовой признак<br>
Далее, мы будем пробовать разные вариант количества отстающих данных и скользщего среднего. Будем вызывать функцию создания признаков, которую написали ранее

```python
def prepare_data(data, target, max_lag, rolling_mean):
    new_df = make_features(data, column=target, 
                           max_lag=max_lag, 
                           rolling_mean_size=rolling_mean)
    train_df, test_df = train_test_split(new_df, test_size=0.1, shuffle=False)
    train = ML_data(train_df)
    test = ML_data(test_df)
    train.df.dropna(inplace=True)
    train.x_y_split(target)
    test.x_y_split(target)
    
    return train, test
```

Проверим работу нашей функции

```python
train, test = prepare_data(df, 'num_orders', 5, 5)
helper.elements_percent('Процент обучающей выборки:', train.df, df_new_features)
helper.elements_percent('Процент тестовой выборки:', test.df, df_new_features)
```

### Обучение моделей

Обучим разные модели и посмотрим, как они спраятся

#### Linear Regressor

```python
top_rmse = 100
top_lags = 0
top_mean = 0
top_fit_time = 0
top_prediction_time = 0

for lag in range(1, 24, 2):
    for rolling_mean in range(5, 51, 5):
        train, test = prepare_data(df, 'num_orders', max_lag=lag, rolling_mean=rolling_mean)
        model = LinearRegression()
        fit_time, prediction_time, rmse_value, predicted = helper.fit_and_predict(model, train, test)
        
        if (rmse_value < top_rmse):
            top_rmse = rmse_value
            top_lags = lag
            top_mean = rolling_mean
            top_fit_time = fit_time
            top_prediction_time = prediction_time
            
print('Лучшая метрика RMSE:', top_rmse)
print('Отстающих данных:', top_lags)
print('Скользящее среднее:', top_mean)
print('Время обучения:', top_fit_time)
print('Время предсказания:', top_prediction_time)
```

#### Lasso

```python
top_rmse = 100
top_lags = 0
top_mean = 0
top_fit_time = 0
top_prediction_time = 0

for lag in range(1, 24, 2):
    for rolling_mean in range(5, 51, 5):
        train, test = prepare_data(df, 'num_orders', max_lag=lag, rolling_mean=rolling_mean)
        model = Lasso(random_state=SEED)
        fit_time, prediction_time, rmse_value, predicted = helper.fit_and_predict(model, train, test)
        
        if (rmse_value < top_rmse):
            top_rmse = rmse_value
            top_lags = lag
            top_mean = rolling_mean
            top_fit_time = fit_time
            top_prediction_time = prediction_time
            
print('Лучшая метрика RMSE:', top_rmse)
print('Отстающих данных:', top_lags)
print('Скользящее среднее:', top_mean)
print('Время обучения:', top_fit_time)
print('Время предсказания:', top_prediction_time)
```

#### Ridge

```python
top_rmse = 100
top_lags = 0
top_mean = 0
top_fit_time = 0
top_prediction_time = 0

for lag in range(1, 24, 2):
    for rolling_mean in range(5, 51, 5):
        train, test = prepare_data(df, 'num_orders', max_lag=lag, rolling_mean=rolling_mean)
        model = Ridge(random_state=SEED)
        fit_time, prediction_time, rmse_value, predicted = helper.fit_and_predict(model, train, test)
        
        if (rmse_value < top_rmse):
            top_rmse = rmse_value
            top_lags = lag
            top_mean = rolling_mean
            top_fit_time = fit_time
            top_prediction_time = prediction_time
            
print('Лучшая метрика RMSE:', top_rmse)
print('Отстающих данных:', top_lags)
print('Скользящее среднее:', top_mean)
print('Время обучения:', top_fit_time)
print('Время предсказания:', top_prediction_time)
```

#### Decision Tree

```python
top_rmse = 100
top_lags = 0
top_mean = 0
top_fit_time = 0
top_prediction_time = 0
top_depth = 0

for lag in range(1, 24, 2):
    for rolling_mean in range(5, 51, 5):
        for depth in range(5, 51, 5):
            train, test = prepare_data(df, 'num_orders', max_lag=lag, rolling_mean=rolling_mean)
            model = DecisionTreeRegressor(max_depth=depth, random_state=SEED)
            fit_time, prediction_time, rmse_value, predicted = helper.fit_and_predict(model, train, test)
            
            if (rmse_value < top_rmse):
                top_rmse = rmse_value
                top_lags = lag
                top_mean = rolling_mean
                top_fit_time = fit_time
                top_prediction_time = prediction_time
                top_depth = depth
            
print('Лучшая метрика RMSE:', top_rmse)
print('Отстающих данных:', top_lags)
print('Скользящее среднее:', top_mean)
print('Глубина:', depth)
print('Время обучения:', top_fit_time)
print('Время предсказания:', top_prediction_time)
```

#### Random Forest

```python
top_rmse = 100
top_lags = 0
top_mean = 0
top_fit_time = 0
top_prediction_time = 0
top_depth = 0
top_estimators = 0

for lag in [1, 5, 15, 24]:
    for rolling_mean in [5, 10, 24, 48]:
        for depth in [5, 25, 50]:
            for est in [50, 60, 70]:
                train, test = prepare_data(df, 'num_orders', max_lag=lag, rolling_mean=rolling_mean)
                model = RandomForestRegressor(max_depth=depth, n_estimators=est, random_state=SEED)
                fit_time, prediction_time, rmse_value, predicted = helper.fit_and_predict(model, train, test)
                
                if (rmse_value < top_rmse):
                    top_rmse = rmse_value
                    top_lags = lag
                    top_mean = rolling_mean
                    top_fit_time = fit_time
                    top_prediction_time = prediction_time
                    top_depth = depth
                    top_estimators = est
            
print('Лучшая метрика RMSE:', top_rmse)
print('Отстающих данных:', top_lags)
print('Скользящее среднее:', top_mean)
print('Глубина:', depth)
print('Оценщиков:', top_estimators)
print('Время обучения:', top_fit_time)
print('Время предсказания:', top_prediction_time)
```

#### LGBM

```python
top_rmse = 100
top_lags = 0
top_mean = 0
top_fit_time = 0
top_prediction_time = 0
top_depth = 0
top_estimators = 0

for lag in [1, 5, 15, 24]:
    for rolling_mean in [5, 10, 24, 48]:
        for depth in [10, 25, 50]:
            for est in [40, 75]:
                train, test = prepare_data(df, 'num_orders', max_lag=lag, rolling_mean=rolling_mean)
                model = LGBMRegressor(max_depth=depth, n_estimators=est, random_state=SEED)
                fit_time, prediction_time, rmse_value, predicted = helper.fit_and_predict(model, train, test)
                
                if (rmse_value < top_rmse):
                    top_rmse = rmse_value
                    top_lags = lag
                    top_mean = rolling_mean
                    top_fit_time = fit_time
                    top_prediction_time = prediction_time
                    top_depth = depth
                    top_estimators = est
            
print('Лучшая метрика RMSE:', top_rmse)
print('Отстающих данных:', top_lags)
print('Скользящее среднее:', top_mean)
print('Глубина:', depth)
print('Оценщиков:', top_estimators)
print('Время обучения:', top_fit_time)
print('Время предсказания:', top_prediction_time)
```

## Анализ моделей

Перейдём к анализу моделей:

- создадим лучшую модель и сравним на графике её предсказания с реальными данными

```python
def feature_importances(model, X, title):
    feature_imp = pd.DataFrame(sorted(zip(model.feature_importances_, 
                                          X.columns)), 
                               columns=['Value','Feature'])
    plt.figure(figsize=(20, 10))
    sns.barplot(x="Value", y="Feature", data=feature_imp.sort_values(by="Value", ascending=False))
    plt.title(title)
    plt.tight_layout()
    plt.show()
```

```python
def lineplot_predictions(title, target, pred):
    fig = plt.figure(figsize=(15,5))
    ax = sns.lineplot(x=target.index, y=target.values, color='red')
    ax = sns.lineplot(x=pred.index, y=pred.values, color='green')
    ax.set(title=title, xlabel='Дата')
    fig.legend(labels=['Исходные данные', 'Предсказания'])
```

### Linear Regressor

```python
lag = 23
rolling_mean = 50
title = 'Линейная регрессия'
train, test = prepare_data(df, 'num_orders', max_lag=lag, rolling_mean=rolling_mean)
model = LinearRegression()
fit_time, prediction_time, rmse_value, predicted = helper.fit_and_predict(model, train, test)
print('Лучшая метрика RMSE:', rmse_value)
print('Отстающих данных:', lag)
print('Скользящее среднее:', rolling_mean)
print('Время обучения:', fit_time)
print('Время предсказания:', prediction_time)
lineplot_predictions(title, test.y, predicted)
```

Линейная регрессия справляется не так хорошо, но проходит по требованию (RMSE <= 48)
